#### CREDIT_DURATION

*Đề xuất tạo thêm features CREDIT_DURATION chứa thông tin về mức độ ổn định và độ dài của khoản vay. Biến này là một biến khá quan trọng để đánh giá khả năng trả nợ của khách hàng.*

```python
bureau_merged['CREDIT_DURATION'] = np.abs(bureau_merged['DAYS_CREDIT'] - bureau_merged['DAYS_CREDIT_ENDDATE'])
plot_numerical_variables(bureau_merged, column_name = 'CREDIT_DURATION')
```

- Có thể thấy rằng phần lớn các khách hàng trong tập dữ liệu có thời gian tín dụng từ 1 đến 7 năm. Số lượng khách hàng có thời gian tín dụng dưới 1 năm và trên 7 năm là tương đối ít, trên 13 năm thì gần như không có. 
- Hầu hết cả hai đối tượng Non-Defaulted và Defaulted đều có tỷ lệ cao nhất ở khoảng 3 năm, trong đó Non-Defaulted cao hơn hẳn so với Defaulted.
- Ở biểu đồ box-plot thì sự phân phối dữ liệu của cả hai đối tượng khá tương đồng, không có sự chênh lệch nào quá lớn.

#### MAX_AMT_OVERDUE_DURATION_RATIO
*Đánh giá mức độ nợ so với thời gian kéo dài của khoản vay. Nếu tỷ lệ này cao, có thể cho thấy nguy cơ trả nợ không đúng hạn.*

```python
bureau_merged['MAX_AMT_OVERDUE_DURATION_RATIO'] = bureau_merged['AMT_CREDIT_MAX_OVERDUE'] / (bureau_merged['CREDIT_DURATION'] + 0.00001)
plot_numerical_variables(bureau_merged, column_name = 'MAX_AMT_OVERDUE_DURATION_RATIO')
```

- Khách hàng có MAX_AMT_OVERDUE_DURATION_RATIO thấp: Đây là nhóm khách hàng có khả năng trả nợ tốt. Nhóm khách hàng này thường thanh toán đúng hạn hoặc thanh toán trễ ít.
- Khách hàng có MAX_AMT_OVERDUE_DURATION_RATIO cao: Đây là nhóm khách hàng có khả năng trả nợ kém. Nhóm khách hàng này thường thanh toán trễ nhiều.

Qua quan sát ở cả 2 biểu đồ distribution và dist-plot thì không có dữ liệu về nhóm khách hàng có MAX_AMT_OVERDUE_DURATION_RATIO cao, vậy nên chúng ta có thể hiểu rằng hầu hết mọi người đều cố gắng trả nợ đúng hạn hoặc là trễ ít.

#### CURRENT_AMT_OVERDUE_DURATION_RATIO
*Cho biết tỷ lệ giữa số tiền nợ hiện tại và thời gian giữa ngày đăng ký và ngày kết thúc của khoản tín dụng. Cung cấp thông tin về mức độ nợ hiện tại so với thời gian kéo dài của khoản vay, giúp đánh giá khả năng trả nợ của khách hàng.*

```python
bureau_merged['CURRENT_AMT_OVERDUE_DURATION_RATIO'] = bureau_merged['AMT_CREDIT_SUM_OVERDUE'] / (bureau_merged['CREDIT_DURATION'] + 0.00001)
plot_numerical_variables(bureau_merged, column_name = 'CURRENT_AMT_OVERDUE_DURATION_RATIO')
```

- Biểu đồ phân phối cho thấy phân bố của CURRENT_AMT_OVERDUE_DURATION_RATIO là lệch phải. Điều này có nghĩa là có nhiều khách hàng có tỷ lệ nợ quá hạn hiện tại thấp hơn so với tỷ lệ nợ quá hạn cao.
- Biểu đồ dist-plot cho thấy so với Non-Defaulted thì Defaulted có tỷ lệ nợ quá hạn hiện tại thấp nhiều hơn so với Non-Defaulted. 
- Non-Defaulted có một số ít là có tỷ lệ nợ quá hạn hiện tại cao. Điều này được thể hiện qua outliers trong biểu đồ box-plot.

#### AMT_OVERDUE_DURATION_LEFT_RATIO
*Cho biết tỷ lệ giữa số tiền nợ hiện tại và thời gian còn lại đến ngày kết thúc của khoản tín dụng. Cung cấp thông tin về mức độ nợ so với thời gian còn lại của khoản vay. Nếu tỷ lệ này cao, có thể cho thấy nguy cơ trả nợ không đúng hạn.*
